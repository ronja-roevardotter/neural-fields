```python
steady_list = [steady[-1], steady[-2], steady[-3], steady[-4], steady[-5], 
               steady[-6], steady[-7], steady[-8], steady[-9], steady[-10]]
bump_list = [bump[-1], bump[-2], bump[-3], bump[-4], bump[-5], 
             bump[-6], bump[-7], bump[-8], bump[-9], bump[-10]]
temporal_list = [temporal[-1], temporal[-2], temporal[-3], temporal[-4], temporal[-5], 
               temporal[-6], temporal[-7], temporal[-8], temporal[-9], temporal[-10]]
```

```python
from scipy.signal import correlate
```

```python
steady_corrs = []
bump_corrs = []
temporal_corrs = []

for i in range(1,10):
    corr = correlate(steady_list[i-1], steady_list[i])
    steady_corrs.append(np.mean(corr))
    print('Correlation-value for steady-state: ', np.mean(corr))
    plt.imshow(corr)
    plt.title('steady')
    plt.show()
    corr = correlate(bump_list[i-1], bump_list[i])
    bump_corrs.append(np.mean(corr))
    print('Correlation-value for bump-state: ', np.mean(corr))
    plt.imshow(corr)
    plt.title('Bump')
    plt.show()
    corr = correlate(temporal_list[i-1], temporal_list[i])
    temporal_corrs.append(np.mean(corr))
    print('Correlation-value for temporal-state: ', np.mean(corr))
    plt.imshow(corr)
    plt.title('Temporal')
    plt.show()
    
```

```python
steady_frequs = np.fft.fftshift(np.fft.fft2(steady[-1]))
bump_frequs = np.fft.fftshift(np.fft.fft2(bump[-1]))
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


steady_norm = sigmoid(steady_frequs)
bump_norm = sigmoid(steady_frequs)

steady_reals = sigmoid(steady_frequs.real)
steady_imags = sigmoid(steady_frequs.imag)

bump_reals = sigmoid(bump_frequs.real)
bump_imags = sigmoid(bump_frequs.imag)
```

```python
np.abs(steady_norm)
```

```python
reals = steady_frequs.real
imags = steady_frequs.imag
abs = np.sqrt(reals**2 + imags**2)
abs
```

```python
fig, ((ax0, ax1), (ax2, ax3), (ax4, ax5), (ax6, ax7)) = plt.subplots(4, 2, figsize=(16,24))

ax0.imshow(steady[-1].T, vmin=0, vmax=1, aspect='auto')
ax1.imshow(bump[-1].T, vmin=0, vmax=1, aspect='auto')
ax0.set_title('stationary activity')
ax1.set_title('spatial activity')


ax2.imshow(np.log(np.abs(steady_norm)), vmin=0, vmax=1, aspect='auto')
ax3.imshow(np.log(np.abs(bump_norm)), vmin=0, vmax=1, aspect='auto')
ax2.set_title(r'$\log(|f_{stat}|)$')
ax3.set_title(r'$\log(|f_{bump}|)$')


ax4.imshow(steady_reals, vmin=0, vmax=1, aspect='auto')
ax5.imshow(bump_reals, vmin=0, vmax=1, aspect='auto')
ax4.set_title(r'$real(f_{stat})$')
ax5.set_title(r'$real(f_{bump})$')


ax6.imshow(steady_imags, vmin=0, vmax=1, aspect='auto')
ax7.imshow(bump_imags, vmin=0, vmax=1, aspect='auto')
ax6.set_title(r'$imag(f_{stat})$')
ax7.set_title(r'$imag(f_{bump})$')
```

```python
steady_abs_diff = np.max(np.abs(steady_frequs)) - np.min(np.abs(steady_frequs))
steady_real_diff = np.max(steady_reals) - np.min(steady_reals)
steady_imag_diff = np.max(steady_imags) - np.min(steady_imags)
steady_abs_diff, steady_real_diff, steady_imag_diff
```

```python
bump_abs_diff = np.max(np.abs(bump_frequs)) - np.min(np.abs(bump_frequs))
bump_real_diff = np.max(bump_reals) - np.min(bump_reals)
bump_imag_diff = np.max(bump_imags) - np.min(bump_imags)
bump_abs_diff, bump_real_diff, bump_imag_diff
```
